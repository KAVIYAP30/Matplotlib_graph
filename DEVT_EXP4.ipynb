```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('/content/Iris.csv')


```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['SepalLengthCm'], label='Sepal Length')
plt.plot(df['SepalWidthCm'], label='Sepal Width')
plt.xlabel('Sample Index')
plt.ylabel('Length/Width (cm)')
plt.title('Sepal Length and Width over Samples')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c=df['Id'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['PetalLengthCm'], bins=20, alpha=0.7, label='Petal Length')
plt.hist(df['PetalWidthCm'], bins=20, alpha=0.7, label='Petal Width')
plt.xlabel('Length/Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length and Width')
plt.legend()
plt.show()
```

```python
species_counts = df['Species'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(species_counts.index, species_counts.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Number of Samples per Species')
plt.show()
```

```python
species_counts = df['Species'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Iris Species')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Group the data by species
grouped = df.groupby('Species')

# Extract sepal length data for each species
setosa_sepal_length = grouped.get_group('Iris-setosa')['SepalLengthCm']
versicolor_sepal_length = grouped.get_group('Iris-versicolor')['SepalLengthCm']
virginica_sepal_length = grouped.get_group('Iris-virginica')['SepalLengthCm']

# Create a box plot
plt.boxplot([setosa_sepal_length, versicolor_sepal_length, virginica_sepal_length], labels=grouped.groups.keys())

# Add labels and a title
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Box Plot of Sepal Length by Species')

# Show the plot
plt.show()






```
